```python
CATALOG = "my_living_index"

# Create schema
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.bronze COMMENT 'Raw ingested data from data.gov.my'")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.silver COMMENT 'Cleaned and typed data'")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.gold COMMENT 'Business-ready analytical tables'")

print(f"Catalog: {CATALOG}")
print(f"Schemas: bronze, silver, gold")
```
